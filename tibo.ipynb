```python
import numpy as np

cam = None
```

```python
import cv2

if cam is not None:
    cam.release()
cam = cv2.VideoCapture(1)
```

```python
from matplotlib import pyplot as plt

plt.imshow(cam.read()[1])
cam.release()
```

```python
while True:
    _, frame = cam.read()
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```
